# Create vector embedding from the JSON object


The new models have a new required input parameter: input_type, that must be set for every API call and include one of the following four values:

<ul>
<li>input_type="search_document": Use this for texts (documents) you want to store in your vector database</li>
<li>input_type="search_query": Use this for search queries to find the most relevant documents in your vector database</li>
<li>input_type="classification": Use this if you use the embeddings as an input for a classification system</li>
<li>input_type="clustering": Use this if you use the embeddings for text clustering</li>
</ul>

Using these input types ensures the highest possible quality for the respective tasks. If you want to use the embeddings for multiple use cases, we recommend using input_type="search_document".

*rif https://txt.cohere.com/introducing-embed-v3/*

# Usiamo LLM cohere to generare l'output TEXT

```python
def generate_text(prompt, temp=0):
  response = co.generate(
    model='command',
    prompt=prompt,
    max_tokens=200,
    temperature=temp)
  return response.generations[0].text
```

```python
# Retrieve the vector
#cursor.execute('select  V from Vector_CV')
cursor.execute('select nome,cognome, V, chunck from Vector_CV')
datav=[]
#crea l'array data per tutti i dati 
data = cursor.fetchall()
#crea l'array datav solo per i vectory 
for n in range(len(data)):
    datav.append(data[n][2])
print (len(datav))
#print (datav)
print (data[0][0],data[0][1],data[0][3] )
print (data[1][0],data[1][1],data[1][3] )
print (data[2][0],data[2][1],data[2][3] )
#print (datav[0])
#print (datav[1])
```

# Funzionalita di solo Oracle vector search evoluzione di oracle search 

```python
from scipy.spatial import distance
import numpy 
vectors=numpy.array(datav)
distances = [distance.cosine(vec_query, numpy.ravel(vector) ) for vector in vectors]
nearest_neighbor_index = distances.index(min(distances))

print(data[nearest_neighbor_index])
print(data[nearest_neighbor_index][0])
print(data[nearest_neighbor_index][1])
print(data[nearest_neighbor_index][3])
```
